### Trying To figure out why Max is 99999

```python
def display_row_by_value(dataframe, column_name, value):
    """
    Function to display rows in a pandas DataFrame based on a specific column value.
    
    Parameters:
    - dataframe: pandas DataFrame object
    - column_name: string, the name of the column to filter on
    - value: the value in the specified column to filter by
    
    Returns:
    - None
    """
    try:
        filtered_rows = dataframe[dataframe['prices'] == 99999.0]
        if not filtered_rows.empty:
            print(f"Rows where 'prices' = 99999.0:")
            print(filtered_rows)
        else:
            print(f"No rows found where 'prices' = 99999.0.")
    except KeyError:
        print(f"Error: Column 'prices' does not exist in the DataFrame.")

# Example usage:
# Assuming 'df' is your pandas DataFrame
# display_row_by_value(df, 'column_name', 'desired_value')
```

## Deriving New Variables for Frequency

```python
# 2. Create new column titled "Busiest Days"
result = []

for value in df_merged["orders_day_of_week"]:
  if value == 0 or value ==1:
    result.append("Busiest days")
  elif value == 3 or value == 4:
    result.append("Slowest days")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
df_merged['busiest_days']=result
```

```python
# 3. Check columns for accuracy
df_merged['busiest_days'].value_counts(dropna = False)
```

```python
# This looks accurate, the values counts are all closer in size because "busiest days" and "slowest days" took values that were previously in "regualryl busy"
```

```python
# 4. Find frequency of hours orders are placed
df_merged['order_hour_placed'].value_counts(dropna = False)
```

```python
# I'm going to put 8 hours in each category (most, average, and fewest orders)
# Most 9-16
# Average 7-8, 17-22
# Fewest 23, 0-6
```

```python
# Using loc( ) to define conditions
df_merged.loc[(df_merged['order_hour_placed'] >= 9) & (df_merged['order_hour_placed'] <= 16), 'busiest_period_of_day'] = 'Most orders'
```

```python
df_merged.loc[(df_merged['order_hour_placed'] == 7) |
    (df_merged['order_hour_placed'] == 8) |
    ((df_merged['order_hour_placed'] >=17) & (df_merged['order_hour_placed'] <=22)), 
    'busiest_period_of_day'] = 'Average orders'
```

```python
df_merged.loc[(df_merged['order_hour_placed'] == 23) | 
    (df_merged['order_hour_placed'] <= 6) , 'busiest_period_of_day'] = 'Fewest Orders'
```

```python
# 5. Print frequency of 'busiest_period_of_day'
df_merged['busiest_period_of_day'].value_counts(dropna = False)
```
