### Importing Data

```python
# Import orders_products_combined pickle
df_ords_prods_combined = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_combined.pkl'))
```

### Drop unnecessary columns from df_prods and df_ords_prods_combined

```python
# Drop 'Unnamed: 0' from df_prods
df_prods = df_prods.drop(columns = ['Unnamed: 0'])
```

```python
df_prods.head()
```

```python
# Drop 'Unnamed: 0' and '_merge' from df_ords_prods_combined
df_ords_prods_combined = df_ords_prods_combined.drop(columns = ['Unnamed: 0','_merge'])
```

```python
df_ords_prods_combined.head()
```

### Merge df_prods and df_ords_prods_combined

```python
df_merged = df_ords_prods_combined.merge(df_prods, on = 'product_id', indicator = True)
df_merged.head()
```

```python
df_merged['_merge'].value_counts()
```

## Exporting df_merged to pickle

```python
df_merged.to_pickle(os.path.join(path, '02 Data','Prepared Data', 'ords_prods_merged.pkl'))
```
