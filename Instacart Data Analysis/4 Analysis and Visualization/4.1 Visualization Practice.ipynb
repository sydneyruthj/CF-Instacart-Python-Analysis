```python
# Import ords_prods_merged.pkl
df = pd.read_pickle(os.path.join(path, '02 Data','Prepared Data', 'combo_7_25_correct.pkl'))
```

### Visualizations

```python
# Create bar chart of order_day_of_week
df['orders_day_of_week'].value_counts().plot.bar()
```

```python
# Reorder
bar = df['orders_day_of_week'].value_counts().sort_index().plot.bar()
```

```python
# Add color
bar = df['orders_day_of_week'].value_counts().sort_index().plot.bar(color = ['red','orange','yellow','green','blue','indigo','violet'])
```

```python
# Exporting the Visualization
bar.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_orders_dow.png'))
```

### Sampling

```python
np.random.seed(4)
dev = np.random.rand(len(df)) <= 0.7
```

```python
dev
```

```python
big = df[dev]
```

```python
small = df[~dev]
```

```python
len(big) + len(small)
```

```python
df_2 = small[['orders_day_of_week','prices']]
```

```python
line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```

```python
# Exporting the Visualization
line.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_dow_sample.png'))
```
