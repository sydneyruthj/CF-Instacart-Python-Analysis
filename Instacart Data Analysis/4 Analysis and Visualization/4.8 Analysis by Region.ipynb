## Analyzing Regions

```python
# Region and 'avg_price'
# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='region', y='avg_price', data=df, palette='viridis')
plt.xlabel('Region')
plt.ylabel('Average Price')
plt.title('Average Price by Region')
plt.show()
```

```python
# Region and 'med_days_prior'
plt.figure(figsize=(10, 6))
sns.barplot(x='region', y='med_days_prior', data=df, palette='viridis')
plt.xlabel('Region')
plt.ylabel('Median Days Since Prior Order')
plt.title('Order Frequency by Region')
plt.show()
```

```python
# Create line plot
sns.lineplot(data=df, x='order_hour_placed', y='order_number', hue='region')

# Add labels and title
plt.xlabel('Hour')
plt.ylabel('Value')
plt.title('Time Orders Placed by Region')

# Show the plot
plt.show()
```

```python
# Region and Spending Flag
contingency_table = pd.crosstab(df['region'], df['spending_flag'])

print(contingency_table)
```

```python
df.groupby('region').agg({'avg_price': ['mean']})
```

```python
# Convert to row proportions
row_proportions_table = contingency_table.div(contingency_table.sum(axis=1), axis=0)

print("Contingency Table with Row Proportions:")
print(row_proportions_table)
```

```python
# Region counts
df['region'].value_counts().plot.bar()
```
